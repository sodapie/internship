```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Noneを空文字列に置き換え
df['article'] = df['article'].apply(lambda x: x if x is not None else '')

# TF-IDF Vectorizerの初期化
vectorizer = TfidfVectorizer()

# TF-IDFの計算
tfidf_matrix = vectorizer.fit_transform(df['article'])

# 結果をDataFrameに変換
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# 各単語の平均TF-IDFスコアを計算
average_tfidf = tfidf_df.mean(axis=0)

# 平均TF-IDFスコアが高い順に並べ替えてトップ50の単語を抽出
top_50_words = average_tfidf.sort_values(ascending=False).head(50)

# トップ50の単語とそのTF-IDFスコアを表示
print(top_50_words)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_histogram(df, column_name, bins=10, title=None, xlabel=None, ylabel='Frequency', color='blue'):
    """
    指定されたカラムのヒストグラムを表示する関数。

    Parameters:
        df (pd.DataFrame): 対象のデータフレーム。
        column_name (str): ヒストグラムを表示したいカラム名。
        bins (int): ヒストグラムのビンの数。デフォルトは10。
        title (str): グラフのタイトル。デフォルトはカラム名。
        xlabel (str): X軸のラベル。デフォルトはカラム名。
        ylabel (str): Y軸のラベル。デフォルトは'Frequency'。
        color (str): ヒストグラムの色。デフォルトは'blue'。
    """

    # カラムが存在するか確認
    if column_name not in df.columns:
        raise ValueError(f"カラム '{column_name}' がデータフレームに存在しません。")

    # ヒストグラムのプロット
    plt.figure(figsize=(10, 6))
    plt.hist(df[column_name].dropna(), bins=bins, color=color, edgecolor='black')

    # タイトルとラベルを設定
    plt.title(title if title else f'{column_name}のヒストグラム')
    plt.xlabel(xlabel if xlabel else column_name)
    plt.ylabel(ylabel)
    
    # グラフを表示
    plt.show()
```

```python
# 時間ごとのツイートの割合
def plot_time_distribution(df, date_column):
    """
    指定された日付・時間カラムを時間単位で分割し、各時間の出現割合を折れ線グラフで表示する関数。

    Parameters:
        df (pd.DataFrame): 対象のデータフレーム。
        date_column (str): 日付・時間が格納されているカラム名。
    """
    
    # 日付・時間カラムをdatetime型に変換
    df[date_column] = pd.to_datetime(df[date_column], format='%d-%b-%Y %I:%M%p')
    
    # 時間単位で分割
    df['Hour'] = df[date_column].dt.hour
    
    # 各時間の出現回数をカウント
    hour_counts = df['Hour'].value_counts().sort_index()
    
    # 各時間の出現割合を計算
    hour_percentages = (hour_counts / hour_counts.sum()) * 100
    
    # 折れ線グラフをプロット
    plt.figure(figsize=(10, 6))
    plt.plot(hour_percentages.index, hour_percentages.values, marker='o')
    
    # グラフのタイトルとラベルを設定
    plt.title(f'{date_column}の時間ごとの出現割合')
    plt.xlabel('時間 (時)')
    plt.ylabel('割合 (%)')
    plt.xticks(hour_percentages.index)  # X軸に時間を表示
    plt.grid(True)
    
    # グラフを表示
    plt.show()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from janome.tokenizer import Tokenizer
from itertools import combinations
from collections import Counter

def create_cooccurrence_network(df, column_name, min_count=2):
    """
    指定されたカラムを使用して共起ネットワークを作成し表示する関数。

    Parameters:
        df (pd.DataFrame): 対象のデータフレーム。
        column_name (str): 共起ネットワークを作成したいカラム名。
        min_count (int): 共起頻度の最小値。この値以上の頻度で共起する単語ペアのみを含める。デフォルトは2。
    """
    
    # Janomeを使ってテキストを形態素解析し、名詞を抽出
    tokenizer = Tokenizer()
    word_list = []

    for text in df[column_name].dropna():
        tokens = [token.base_form for token in tokenizer.tokenize(text) if token.part_of_speech.startswith('名詞')]
        word_list.append(tokens)
    
    # 共起ペアを抽出
    cooccurrence_pairs = []
    for words in word_list:
        cooccurrence_pairs.extend(combinations(words, 2))
    
    # 共起ペアの頻度をカウント
    cooccurrence_counts = Counter(cooccurrence_pairs)

    # 共起頻度がmin_count以上のペアのみをネットワークに追加
    G = nx.Graph()
    for (word1, word2), count in cooccurrence_counts.items():
        if count >= min_count:
            G.add_edge(word1, word2, weight=count)
    
    # ネットワークを描画
    plt.figure(figsize=(12, 12))
    pos = nx.spring_layout(G, k=0.5)  # ノードの位置を決定
    weights = [G[u][v]['weight'] for u, v in G.edges()]
    nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, font_size=10, edge_color=weights, edge_cmap=plt.cm.Blues, width=2)
    plt.title(f'{column_name} の共起ネットワーク')
    plt.show()
```
